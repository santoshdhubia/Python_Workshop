```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
from scipy.misc import imread


%matplotlib inline
```

# Get the most liked and most retweet tweets

```python
fav_max = np.max(df['Likes'])
rt_max  = np.max(df['RTs'])
```

```python
rt_max
```

```python
fav = df[df.Likes == fav_max].index[0]
rt  = df[df.RTs == rt_max].index[0]
```

```python
fav
```

```python
print("The tweet with highest likes is: \n{}",df['Tweets'][fav])
print("Number of likes: {}".format(fav_max))
print("{} characters.\n".format(df['len'][fav]))
```

```python
print("The tweet with highest retweets is: \n{}".format(df['Tweets'][rt]))
print("Number of retweets: {}".format(rt_max))
print("{} characters.\n".format(df['len'][rt]))
```

# Wordclod from tweets

```python
Tweets_Combined = " ".join(df['Tweets'].values.astype(str))
```

```python
wcloud = WordCloud(background_color="black", stopwords=STOPWORDS)
wcloud.generate(Tweets_Combined)
```

```python
plt.imshow(wcloud)
plt.axis("off")
```

```python
Tweets_nourl_notags= " ".join([Tweets_Combined for Tweets_Combined in Tweets_Combined.split()
                            if 'http' not in Tweets_Combined
                                and not Tweets_Combined.startswith('@')
                                and Tweets_Combined != 'RT'
                            ])
```

```python
wcloud1 = WordCloud(background_color="white", stopwords=STOPWORDS)
wcloud1.generate(Tweets_nourl_notags)
```

```python
plt.imshow(wcloud1)
plt.axis('off')
```

```python
Tweet_mask = imread(r"C:\Users\santosh.d\Desktop\twitter_mask.png", flatten=True)
```

```python
#wcloud1 = WordCloud(background_color="white", stopwords=STOPWORDS, mask=Tweet_mask)
wcloud2 = WordCloud(background_color="white", stopwords=STOPWORDS, mask=Tweet_mask)
wcloud2.generate(Tweets_nourl_notags)
```

```python
plt.figure( figsize=(20,10))
plt.imshow(wcloud2)
plt.axis('off')
plt.savefig(r"C:\Users\santosh.d\Desktop\wordcloud.png",dpi=300)
plt.show()
```
